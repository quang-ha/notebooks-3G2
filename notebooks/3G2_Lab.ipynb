## 3. DYNAMICS OF THE VOCAL FOLDS

For full discription of the terminology used in this report, please refer to the lab handout. 

### 3.1 KINEMATICS DESCRIPTION

In order to represent the phase difference between the subglotal and supraglotal openings, the following equations representing the width of these openings $x_1$ and $x_2$ are presented below:

$$
x_1 = x_{01} + f(t+\tau)\, \, \mathrm{and}\, \, x_2 = x_{02} + f(t-\tau)
$$

with the value of $\tau$ being:

$$
\tau = \frac{T}{2c_{w}}
$$

From this, the mean displacement of each fold can be deduced as:

$$
\epsilon = \frac{1}{2}[f(t+\tau) + f(t-\tau)]
$$

yet using Tayler expansion:
$$
f(t+\tau) = f(t) + \tau f^{'}(t) + O(\tau^{2})
$$

$$
f(t-\tau) = f(t) - \tau f^{'}(t) + O(\tau^{2})
$$

leading to $\epsilon = f(t)$ in the limit of small value of time difference $\tau$ (which makes higher order term in the above equations negligible). Substituting this into the equation for the openings of the upper and the lower part of the vocal holds lead to:

$$
x_{1} = x_{01} + \epsilon + \tau \dot{\epsilon}
$$

$$
x_{2} = x_{02} + \epsilon - \tau \dot{\epsilon}
$$

Finally, the equation governing the subglotal and supraglotal areas are presented below, by utilising the new equation for $x_{1}$ and $x_{2}$:

$$
a_{1} = 2W(x_{01} + \epsilon + \tau \dot{\epsilon})
$$

$$
a_{2} = 2W(x_{02} + \epsilon - \tau \dot{\epsilon})
$$


### 3.2 FORCE BALANCE

For full derivation of the following equations please refer to the lab handout. 

The pressure force on the fold tissue follows the second order differential equation:

$$
M \ddot{\epsilon} + B \dot{\epsilon} + K{\epsilon} = P_{t} \Big( \frac{\Delta x_0 + 2 \tau \dot{\epsilon}}{x_{01} + \epsilon + \tau \dot{\epsilon}} \Big)
$$

A suitable expression to transform this second order ODE into two first order ODE is presented below:

For the mucosal model, the following first order ODE system is found:

$$
f \Big[\begin{pmatrix}
\epsilon \\
\dot{\epsilon} \\
\end{pmatrix} \Big]= \begin{pmatrix}
\dot{\epsilon} \\
\frac{1}{M}\Bigg( P_L \Big[ \frac{\Delta x_0 + 2\tau \dot{\epsilon}}{x_{01}+\epsilon+\tau \dot{\epsilon}} \Big] - B \dot{\epsilon} - K \epsilon \Bigg) \\ 
\end{pmatrix}
$$

As seen from the equation, there are now more parameters in the equation. However, the real parameters controlling the natural frequency of the system are still K and M. This is because the right hand side does not contain any $\ddot{\epsilon}$ or $\epsilon$ terms, which are responsible for the natural frequency of the system. Therefore, the natural frequency of the system remains: 

$$
\omega_{n} = \sqrt{\frac{K}{M}} 
$$

where as the damping becomes more complicated compared to a simple mass-spring system, which is considered below.

### 3.3 NUMERICAL SOLUTION

#### 3.3.1 OSCILATTING MODEL

The ODE system in the provided code can be rewritten as below:

$$
f \Big[\begin{pmatrix}
x \\
\dot{x} \\
\end{pmatrix} \Big]= \begin{pmatrix}
\dot{x} \\
\frac{-Kx-B\dot{x}}{M} \\
\end{pmatrix}
$$

which is simply equivalent to a second order equation for a vibration system:

$$
M\ddot{x} + B\dot{x} + Kx = 0
$$

The natural frequency and damping of the system is:

$$
\omega_{n} = \sqrt{\frac{K}{M}} \\
\zeta = \frac{B}{2\sqrt{KM}}
$$

The script below include an interactive widget that allows visualising the effect of changing M, K, B on the final result of the system. 

```python
from scipy.integrate import odeint

# Import the interactive function from IPython widgets
from IPython.html.widgets import interact
from IPython.html import widgets

# Initial parameters
p = {'P1': 1500,'M': 0.45,'K': 250000,'B': 100,'x01': 0.1,'dx0': 0.01,'tau': 0.002}
```

```python
# Second order differential equation numerical solver
def solver_oscil(x, t):
    return [x[1], (-p['K']*x[0]-p['B']*x[1])/p['M']]
```

Now use numerical integration to solve the system.

```python
# Generating time step
t = np.linspace(0, 0.1, 100000)
```

```python
   # Initial condition and solve the system
x0 = [1e-6, 0]
result = odeint(solver_oscil, x0, t)
x, dx = result.T
   
print "Natural frequency: ",np.sqrt(p['K']/p['M'])
print "Damping ratio: ",p['B']/(2*np.sqrt(p['K']*p['M']))
fig, axes = plt.subplots(1, 2, figsize=(10,5))
axes[0].plot(t,x)
axes[0].grid()
axes[0].set_title('x')
axes[1].plot(t,dx,'r')
axes[1].grid()
axes[1].set_title('xdot')
```

#### 3.3.2 MUCOSAL MODEL

The script is now rewritten to solve the mucosal model quoted from before. 

```python
# Define solver for the mucosal system
def solver_mus(e, t):
    temp = (1/p['M'])*(p['P1']*(p['dx0']+2*p['tau']*e[1])/(p['x01']+e[0]+p['tau']*e[1]) \
                       - p['B']*e[1] - p['K']*e[0])
    return [e[1], temp]
```

```python
# Initial condition and solve the system
e0 = [1e-6, 0]
result = odeint(solver_mus, e0, t)
    
e, de = result.T
fig, axes = plt.subplots(1, 3, figsize=(15,5))
axes[0].plot(t,e)
axes[0].grid()
axes[0].set_title('epsilon')
axes[1].plot(t,de,'g')
axes[1].grid()
axes[1].set_title('epsilon_dot')
axes[2].plot(e, de,'r')
axes[2].grid()
axes[2].set_title('epsilon against epsilon_dot')
```

By changing the value of $P_{c}$, one can obtain the certain value where osciallition with no damping occurs. Overcrossing this value of $P_{c}$ will cause exponential growth of the oscillation system. 

```python
# Redefine Pc
p['P1'] = 2660

result = odeint(solver_mus, e0, t)
    
e, de = result.T
fig, axes = plt.subplots(1, 3, figsize=(15,5))
axes[0].plot(t,e)
axes[0].grid()
axes[0].set_title('epsilon')
axes[1].plot(t,de,'g')
axes[1].grid()
axes[1].set_title('epsilon_dot')
axes[2].plot(e, de,'r')
axes[2].grid()
axes[2].set_title('epsilon against epsilon_dot')
```

Examining the equation quoted in 5.2 will help explaining the phenomenom. At the denominator of the right hand side, we have:

$$
x_{01} + \epsilon + \tau \dot{\epsilon}
$$

It is obvious that when $\epsilon + \tau \dot{\epsilon} = -x_{01}$, the fraction on the right hand side will blow up to infinity, due to division by 0, causing the system to become unstable. This will cause huge problem when doing the numerical integration as seen from the graph. In Python Scipy module, the numerical integration simply give up and immediately equating the rest of the values of $\dot{\epsilon}$ to 0. Whereas scripts running in MATLAB would still produce some results but with spiky peaks represent the infinity points of the obtained solution. 

This is due to the fact that there is no terms in the equation to limit the mean displacement of the vocal cord. In reality, the vocal cord cannot close anyfurther than its initial width $x_{01}$. The numerical ODE solver does not have that particular constraint, hence when this happens the solution blows up. The problem can be solved either by truncated the data at some maximum/minimum value, or introduce an additional damping function to the system. The latter choice turns out to be more logical. As the folds start touching each other, one would expect they will dissipate greater amount of energy.
